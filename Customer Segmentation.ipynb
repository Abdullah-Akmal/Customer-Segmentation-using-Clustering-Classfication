```python
selected_features = combined_df[
    ['quantity', 'unit_price', 'total_price', 'base_price',
     'price_tier','gender', 'payment_method', 'region_x', 'loyalty_tier', 'category']
]

selected_features.head()
```

```python
numerical_cols = ['quantity', 'unit_price', 'total_price', 'base_price']
nominal_cols = ['gender', 'payment_method', 'region_x', 'category']
ordinal_cols = ['price_tier', 'loyalty_tier']
```

```python
from sklearn.impute import SimpleImputer

numerical_transformer = SimpleImputer(strategy='median')
categorical_transformer = SimpleImputer(strategy='most_frequent')
```

```python
price_tier_order = ['Budget', 'Mid', 'Premium', 'Luxury']
loyalty_tier_order = ['Bronze','Silver', 'Gold'] # Assuming Gold > Silver

ordinal_categories = [price_tier_order, loyalty_tier_order]
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
one_hot_encoder = OneHotEncoder(handle_unknown='ignore', drop='first')
```

```python
ordinal_encoder = OrdinalEncoder(categories=ordinal_categories)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('ohe', Pipeline(steps=[('imputer', SimpleImputer(strategy='most_frequent')),
                               ('onehot', one_hot_encoder)]), nominal_cols),
        ('ord', Pipeline(steps=[('imputer', SimpleImputer(strategy='most_frequent')),
                               ('ordinal', ordinal_encoder)]), ordinal_cols)
    ],
    remainder='passthrough' # Keep other columns (if any)
)
```

```python
df_encoded = preprocessor.fit_transform(selected_features)
feature_names = preprocessor.get_feature_names_out()
df_processed = pd.DataFrame(df_encoded, columns=feature_names)

df_processed.head()
```

```python
numerical_cols = ['quantity', 'unit_price', 'total_price', 'base_price']
nominal_cols = ['gender', 'payment_method', 'region_x', 'category']
ordinal_cols = ['price_tier', 'loyalty_tier']
```

```python
numerical_transformer = SimpleImputer(strategy='median')
categorical_transformer = SimpleImputer(strategy='most_frequent')
```

```python
one_hot_encoder = OneHotEncoder(handle_unknown='ignore', drop='first')
```

```python
ordinal_encoder = OrdinalEncoder(categories=ordinal_categories)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('ohe', Pipeline(steps=[('imputer', SimpleImputer(strategy='most_frequent')),
                               ('onehot', one_hot_encoder)]), nominal_cols),
        ('ord', Pipeline(steps=[('imputer', SimpleImputer(strategy='most_frequent')),
                               ('ordinal', ordinal_encoder)]), ordinal_cols)
    ],
    remainder='passthrough' # Keep other columns (if any)
)
```

```python
df_encoded = preprocessor.fit_transform(selected_features)
feature_names = preprocessor.get_feature_names_out()
df_processed = pd.DataFrame(df_encoded, columns=feature_names)

df_processed.head()
```

```python
df_processed.rename(columns={'remainder__labels':'labels'}, inplace=True)
df_processed.columns
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,classification_report
```

```python
X = selected_features.drop(columns=['labels'])
Y=df_processed['labels']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
rf= RandomForestClassifier(n_estimators=100,max_features='sqrt',random_state=42)
```

```python
full_pipeline = Pipeline([
    ('preprocessor', preprocessor),  # Step 1: Translate data
    ('model', rf)                    # Step 2: Train model
])
full_pipeline.fit(X_train, Y_train)
```

```python
y_predict=full_pipeline.predict(X_test)
```

```python
cm= confusion_matrix(Y_test,y_predict)
print(cm)
```

```python
ac= accuracy_score(Y_test,y_predict)
print(ac)
```

```python
cr=classification_report(Y_test,y_predict)
print(cr)
```

```python
import gradio as gr

#Load the saved pipeline
pipeline = joblib.load(r'G:\Other computers\My Laptop\Education and Bootcamp\COMSATS\5th Semester - Fall 2025\MGT - 244 Business Application using Machine Learning\Project\model.pkl')

#Define the logic to process inputs and predict
def predict_cluster(quantity, unit_price, base_price, gender, payment_method, region, category, loyalty_tier):
    
    # Calculate derived features (logic taken from your notebook)
    total_price = quantity * unit_price
    
    # Calculate price_tier based on base_price quartiles from your dataset statistics
    # Quartiles from df.describe(): 25%=14.67, 50%=23.05, 75%=31.96
    if base_price <= 14.67:
        price_tier = "Budget"
    elif base_price <= 23.05:
        price_tier = "Mid"
    elif base_price <= 31.96:
        price_tier = "Premium"
    else:
        price_tier = "Luxury"

    # Create a DataFrame matching the training data structure exactly
    input_df = pd.DataFrame({
        'quantity': [quantity],
        'unit_price': [unit_price],
        'total_price': [total_price],
        'base_price': [base_price],
        'gender': [gender],
        'payment_method': [payment_method],
        'region_x': [region],  # Note: Your model uses 'region_x'
        'category': [category],
        'price_tier': [price_tier],
        'loyalty_tier': [loyalty_tier]
    })

    # Make prediction using the pipeline (preprocessor + model)
    prediction = pipeline.predict(input_df)[0]
    
    # Map cluster labels to meaningful names (optional, based on your analysis)
    cluster_names = {
        0: "Cluster 0 (e.g., High Value / Loyal)",
        1: "Cluster 1 (e.g., Budget / Occasional)",
        2: "Cluster 2 (e.g., Mid-Range / New)"
    }
    
    return f"Predicted Customer Segment: {cluster_names.get(prediction, prediction)}"

#Define the Gradio Interface
iface = gr.Interface(
    fn=predict_cluster,
    inputs=[
        gr.Slider(minimum=1, maximum=10, step=1, label="Quantity"),
        gr.Number(label="Unit Price", value=20.0),
        gr.Number(label="Base Price (Cost)", value=15.0),
        gr.Dropdown(choices=["Male", "Female", "Other"], label="Gender"),
        gr.Dropdown(choices=["Credit Card", "PayPal", "Bank Transfer"], label="Payment Method"),
        gr.Dropdown(choices=["North", "South", "East", "West", "Central"], label="Region"),
        gr.Dropdown(choices=["Cleaning", "Outdoors", "Kitchen", "Personal Care", "Storage"], label="Category"),
        gr.Dropdown(choices=["Bronze", "Silver", "Gold"], label="Loyalty Tier"),
    ],
    outputs="text",
    title="Customer Segmentation Predictor",
    description="Enter customer and purchase details to predict their segment."
)

#Launch the app
iface.launch()
```

```python
import streamlit as st


# 1. Load the saved pipeline
# @st.cache_resource speeds up reloading by caching the model
@st.cache_resource
def load_pipeline():
    return joblib.load(r'G:\Other computers\My Laptop\Education and Bootcamp\COMSATS\5th Semester - Fall 2025\MGT - 244 Business Application using Machine Learning\Project\model.pkl')

pipeline = load_pipeline()

# 2. App Layout
st.title(" Customer Segmentation Predictor")
st.write("Enter the customer transaction details below to predict their segment.")

# Create columns for a better layout
col1, col2 = st.columns(2)

with col1:
    st.subheader("Transaction Details")
    quantity = st.slider("Quantity", min_value=1, max_value=20, value=1)
    unit_price = st.number_input("Unit Price ($)", value=20.0, step=0.5)
    base_price = st.number_input("Base Price/Cost ($)", value=15.0, step=0.5)
    
with col2:
    st.subheader("Customer Info")
    gender = st.selectbox("Gender", ["Male", "Female", "Other"])
    loyalty_tier = st.selectbox("Loyalty Tier", ["Bronze", "Silver", "Gold"])
    payment_method = st.selectbox("Payment Method", ["Credit Card", "PayPal", "Bank Transfer", "Cash"])

st.subheader("Product Details")
col3, col4 = st.columns(2)
with col3:
    region = st.selectbox("Region", ["North", "South", "East", "West", "Central"])
with col4:
    category = st.selectbox("Category", ["Cleaning", "Outdoors", "Kitchen", "Personal Care", "Storage", "Electronics", "Clothing"])

# 3. Prediction Logic
if st.button("Predict Segment", type="primary"):
    
    # Calculate derived features automatically
    total_price = quantity * unit_price
    
    # Logic for price_tier based on your notebook's quartiles
    if base_price <= 14.67:
        price_tier = "Budget"
    elif base_price <= 23.05:
        price_tier = "Mid"
    elif base_price <= 31.96:
        price_tier = "Premium"
    else:
        price_tier = "Luxury"

    # Create DataFrame (names must match training data exactly)
    input_df = pd.DataFrame({
        'quantity': [quantity],
        'unit_price': [unit_price],
        'total_price': [total_price],
        'base_price': [base_price],
        'gender': [gender],
        'payment_method': [payment_method],
        'region_x': [region],  # Mapped to match the notebook's 'region_x'
        'category': [category],
        'price_tier': [price_tier],
        'loyalty_tier': [loyalty_tier]
    })

    # Predict
    try:
        prediction = pipeline.predict(input_df)[0]
        
        # Display Result with custom styling
        st.markdown("---")
        st.header("Prediction Result")
        
        # You can customize these descriptions based on your cluster analysis
        if prediction == 0:
            st.success(f"**Cluster 0** (e.g., High Value Customers)")
        elif prediction == 1:
            st.info(f"**Cluster 1** (e.g., Occasional Buyers)")
        elif prediction == 2:
            st.warning(f"**Cluster 2** (e.g., New/Churn Risks)")
        else:
            st.write(f"Cluster {prediction}")
            
    except Exception as e:
        st.error(f"An error occurred during prediction: {e}")

# Sidebar info
st.sidebar.info("This app uses a Random Forest Classifier trained on sales data to segment customers.")
```
